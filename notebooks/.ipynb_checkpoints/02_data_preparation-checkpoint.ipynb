```python
# Cell 1: Setup and Path Configuration
import os
import numpy as np
import pandas as pd
import geopandas as gpd
import rasterio
from rasterio.windows import Window
from shapely.geometry import Point
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
from tqdm.notebook import tqdm

# Configure paths (relative to notebook location)
base_dir = os.path.abspath(os.path.join('..'))  # Goes up to project root
data_dir = os.path.join(base_dir, 'data')
output_dir = os.path.join(base_dir, 'outputs')

# Input files
filtered_images_path = os.path.join(output_dir, 'reports', 'filtered_images.geojson')
landcover_path = os.path.join(data_dir, 'raw', 'tif', 'worldcover_bbox_delhi_ncr_2021.tif')

# Output files
train_path = os.path.join(data_dir, 'processed', 'train_samples.geojson')
test_path = os.path.join(data_dir, 'processed', 'test_samples.geojson')

# Create processed directory if not exists
os.makedirs(os.path.join(data_dir, 'processed'), exist_ok=True)

print("Configuration verified:")
print(f"Project root: {base_dir}")
print(f"Landcover path: {landcover_path}")
```

```python
# Cell 2: Load and Verify Q1 Output
def load_q1_output(path):
    """Robust loading of filtered images"""
    if not os.path.exists(path):
        raise FileNotFoundError(
            f"Q1 output not found at {path}\n"
            "Please ensure:\n"
            "1. 01_spatial_analysis.ipynb was run completely\n"
            "2. The file was saved to outputs/reports/"
        )
    
    filtered = gpd.read_file(path)
    print(f"Loaded {len(filtered)} filtered images")
    print("Sample records:")
    return filtered

try:
    filtered = load_q1_output(filtered_images_path)
    display(filtered.head(2))
    
except Exception as e:
    print(f"Error: {str(e)}")
    print("\nDirectory contents of outputs/reports/:")
    print(os.listdir(os.path.join(output_dir, 'reports')))
    raise
```

```python
# Cell 3: Patch Extraction from Landcover TIFF
def extract_patch(image_row, patch_size=128):
    """Extract landcover patch centered at image coordinates"""
    try:
        with rasterio.open(landcover_path) as src:
            x, y = image_row.geometry.x, image_row.geometry.y
            row, col = src.index(x, y)
            patch = src.read(
                1,
                window=Window(col-patch_size//2, row-patch_size//2, patch_size, patch_size),
                boundless=True,
                fill_value=0  # ESA NoData=0
            )
        return patch
    except Exception as e:
        print(f"Error processing {image_row.get('filename', 'unknown')}: {str(e)}")
        return None

# Test extraction
sample_patch = extract_patch(filtered.iloc[0])
plt.imshow(sample_patch)
plt.colorbar()
plt.title("Sample Landcover Patch (128×128)")
plt.savefig(os.path.join(output_dir, 'figures', 'sample_patch.png'))
plt.show()
```

```python
# Cell 4: Label Assignment and Class Mapping
# ESA WorldCover 2021 Class Mapping
CLASS_MAPPING = {
    10: 'Tree Cover',
    20: 'Shrubland',
    30: 'Grassland',
    40: 'Cropland',
    50: 'Built-up',
    60: 'Bare/Sparse Vegetation',
    70: 'Snow/Ice',
    80: 'Water',
    90: 'Wetlands',
    95: 'Moss/Lichen',
    100: 'No Data'
}

def assign_label(patch):
    """Calculate dominant class label"""
    if patch is None:
        return 'Invalid'
    
    values = patch.flatten()
    values = values[values != 0]  # Remove NoData
    
    if len(values) == 0:
        return 'No Data'
    
    counts = np.bincount(values)
    return CLASS_MAPPING.get(np.argmax(counts), 'Unknown')

# Process all images
filtered['landcover_patch'] = filtered.apply(extract_patch, axis=1)
filtered['label'] = filtered['landcover_patch'].apply(assign_label)

# Save intermediate results
filtered.to_file(os.path.join(output_dir, 'reports', 'labeled_images.geojson'), driver='GeoJSON')
print("Label distribution:")
print(filtered['label'].value_counts())
```

```python
# Cell 5: Edge Case Handling
# Remove invalid/no-data samples
valid_samples = filtered[~filtered['label'].isin(['Invalid', 'No Data'])]

# Analyze mixed patches
def calculate_mixedness(patch):
    unique, counts = np.unique(patch[patch != 0], return_counts=True)
    if len(unique) == 0:
        return 0
    return 1 - (np.max(counts) / np.sum(counts))

valid_samples['mixedness'] = valid_samples['landcover_patch'].apply(calculate_mixedness)
print(f"Removed {len(filtered)-len(valid_samples)} invalid/no-data samples")
print(f"{len(valid_samples[valid_samples['mixedness'] > 0.7])} highly mixed patches")
```

```python
# Cell 6: Train-Test Split
train_df, test_df = train_test_split(
    valid_samples,
    test_size=0.4,
    random_state=42,
    stratify=valid_samples['label']
)

# Save splits
train_df.to_file(train_path, driver='GeoJSON')
test_df.to_file(test_path, driver='GeoJSON')

print(f"Train samples: {len(train_df)}")
print(f"Test samples: {len(test_df)}")
print(f"Saved to:\n- {train_path}\n- {test_path}")
```

```python
# Cell 7: Class Distribution Visualization
plt.figure(figsize=(14, 6))
sns.countplot(
    data=pd.concat([
        train_df.assign(split='Train'),
        test_df.assign(split='Test')
    ]),
    x='label',
    hue='split',
    order=train_df['label'].value_counts().index
)
plt.title("Class Distribution (60/40 Stratified Split)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, 'figures', 'class_distribution.png'))
plt.show()

# Imbalance analysis
class_dist = train_df['label'].value_counts(normalize=True)
print("\nClass Distribution (Train):")
print(class_dist)
```
